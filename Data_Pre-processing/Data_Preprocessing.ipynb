```python
df.drop(["ca", "thal"], axis=1, inplace=True)
```

```python
num_cols = ["oldpeak", "trestbps", "thalch", "chol"]
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
for col in ["fbs", "exang", "restecg", "slope"]:
    df[col] = df[col].fillna(df[col].mode()[0])
```
